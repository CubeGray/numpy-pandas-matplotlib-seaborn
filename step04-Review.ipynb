```python
df22 = pd.DataFrame(columns=['지역명','분양가격(평)','연도월'])
df22
```

```python
for key,value in df2.iterrows():
        df22 = df22.append({'지역명':value.지역명,'분양가격(평)':value['분양가격(평)'],'연도월':value['연도월']}, ignore_index=True)
```

```python
df22['분양가격(평)'] = pd.to_numeric(df22['분양가격(평)'], errors = 'coerce')
df22['분양가격(평)'].fillna(df22['분양가격(평)'].mean(),inplace=True)
```

```python
count=0
for key, value in df22.iterrows():
    if value.지역명=='서울':
        df22.loc[count,'소속지역']='서울'
    elif value.지역명=='세종':
        df22.loc[count,'소속지역']='세종'
    elif value.지역명=='경기':
        df22.loc[count,'소속지역']='경기'
    elif value.지역명=='부산' or  value.지역명 == '대구' or  value.지역명 == '인천' or  value.지역명 == '광주' or value.지역명 == '대전' or  value.지역명 == '울산':
        df22.loc[count,'소속지역']='광역시'
    elif value.지역명=='강원' or value.지역명=='충북' or value.지역명=='충남' or value.지역명=='전북' or value.지역명=='전남' or value.지역명== '경북' or value.지역명== '경남' or value.지역명=='제주':
        df22.loc[count,'소속지역']='지방'
    count+=1
```
